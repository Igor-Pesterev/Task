```python
# Используемые библиотеки:
import pandas as pd
import numpy as np
import scipy.stats as st


# Создание DataFrame и удаление столбца 'Type', для дальнейшего анализа данных
data = pd.read_csv('C:/Users/admin/Downloads/dataset_Facebook.csv', sep = ';')
data.drop('Type', axis=1, inplace=True)


# DataFrame исходный
data_new = pd.read_csv('C:/Users/admin/Downloads/dataset_Facebook.csv', sep = ';')      

#DataFrame по типу Photo:
data_Photo = pd.DataFrame(columns=list(data_new))


#DataFrame по типу Status:
data_Status = pd.DataFrame(columns=list(data_new))


#DataFrame по типу Video:
data_Video = pd.DataFrame(columns=list(data_new))


#DataFrame по типу Link:
data_Link = pd.DataFrame(columns=list(data_new))


#Цикл который выделяет все значения типа Photo:
for i in range(0,len(data_new)):
    if data_new['Type'][i] == 'Photo':
        data_Photo.loc[i] = data_new.loc[i]


#Цикл который выделяет все значения типа Status:
for i in range(0,len(data_new)):
    if data_new['Type'][i] == 'Status':
        data_Status.loc[i] = data_new.loc[i]


#Цикл который выделяет все значения типа Video:
for i in range(0,len(data_new)):
    if data_new['Type'][i] == 'Video':
        data_Video.loc[i] = data_new.loc[i]


#Цикл который выделяет все значения типа Link:
for i in range(0,len(data_new)):
    if data_new['Type'][i] == 'Link':
        data_Link.loc[i] = data_new.loc[i]


#Метод, организующий расчет числовых характеристик: 
#(медиана, выборочное среднее, мода, минимальное и максимальное значения) для всего массива данных
def num_char(data):
   
    mean = np.mean(data)  # Выборочное среднее
    minimum = np.min(data)  # Минимальное значение
    maximum = np.max(data)  # Максимальное значение
    mode = st.mode(data)  # Мода
    median = np.nanmedian(data)  # Медиана
        
    return (mean, minimum, maximum, mode, median)


#Метод, организующий расчет числовых характеристик для типа Photo
def num_char_Photo(data_Photo):
   
    mean = np.mean(data_Photo) 
    minimum = np.min(data_Photo)
    maximum = np.max(data_Photo)
    mode = st.mode(data_Photo)  
    median = np.median(data_Photo)
        
    return (mean, minimum, maximum, mode, median)


#Метод, организующий расчет числовых характеристик для типа Status
def num_char_Status(data_Status):
   
    mean = np.mean(data_Status) 
    minimum = np.min(data_Status)
    maximum = np.max(data_Status)
    mode = st.mode(data_Status)  
    median = np.median(data_Status)
        
    return (mean, minimum, maximum, mode, median)


#Метод, организующий расчет числовых характеристик для типа Video
def num_char_Video(data_Video):
   
    mean = np.mean(data_Video) 
    minimum = np.min(data_Video)
    maximum = np.max(data_Video)
    mode = st.mode(data_Video)  
    median = np.median(data_Video)
        
    return (mean, minimum, maximum, mode, median)


#Метод, организующий расчет числовых характеристик для типа Link
def num_char_Link(data_Link):
   
    mean = np.mean(data_Link) 
    minimum = np.min(data_Link)
    maximum = np.max(data_Link)
    mode = st.mode(data_Link)  
    median = np.median(data_Link)
        
    return (mean, minimum, maximum, mode, median)


#Нахождение максимумов параметров: like, comment, share
    
max_like = np.max(data['like'])  
max_comment = np.max(data['comment'])
max_share = np.max(data['share'])



# Вывод
print('Числовые характеристики массива данных: ', num_char(data)) 
print('Числовые характеристики массива данных по типу Photo: ', num_char_Photo(data_Photo))
print('Числовые характеристики массива данных по типу Status: ', num_char_Status(data_Status))
print('Числовые характеристики массива данных по типу Video: ', num_char_Video(data_Video))
print('Числовые характеристики массива данных по типу Link: ', num_char_Link(data_Link))
print('Максимум по трем параметрам: ', 'Like: ', max_like, 'Comment: ', max_comment, 'Share: ', max_share)





#Найти самый популярный объект в выборке
# Я нахожу самый популярный элемент по трем пораметрам: like, share, comment. 
#Нашел максимум этих трех параметров и сделал вывод, что самы популярный объект в выборке это:
#130791	Photo	2	7	3	5	1	180480	319133	8072	4010	6242	108752	51456	3316	372	5172	790	6334
```
